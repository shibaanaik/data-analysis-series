```python
#What is the correlation between original price, discount, and final price?
numerical=df[['Price (Rs.)','Discount (%)','Final_Price(Rs.)']]
correlation=numerical.corr()
print(correlation)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation between Price, Discount, and Final Price")
plt.show()
```

From the correlation matrix, we observe a strong positive correlation between original price and final price (0.936), indicating that higher-priced items tend to remain more expensive even after discounts. There's a moderate negative correlation (-0.312) between discount percentage and final price, meaning greater discounts typically lower the final price. However, the discount percentage is almost uncorrelated with the original price, suggesting discounts are applied independently of product price.

```python
#How does the number of purchases vary month by month?
df['Purchase_Date']=pd.to_datetime(df['Purchase_Date'],dayfirst=True)
df['month']=df['Purchase_Date'].dt.strftime('%b')
monthly_purchases = df.groupby('month').size()
print (monthly_purchases)
```

```python
import matplotlib.pyplot as plt

monthly_purchases.index = monthly_purchases.index.astype(str)
plt.figure(figsize=(10, 5))
monthly_purchases.plot(kind='bar', color='mediumseagreen')
plt.title('Monthly Purchase Count')
plt.xlabel('Month')
plt.ylabel('Number of Purchases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
